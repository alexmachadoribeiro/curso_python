# **IDEs para Python**
---

## Pré-requisito da aula

- Introdução ao Python
---

Na aula de **Introdução ao Python** vimos um pouco sobre **IDEs**, também conhecidos como **Ambiente de Desenvolvimento Integrado**. Na aula de hoje, vamos aprender um pouco mais a fundo sobre eles.

## Visual Studio Code (VSCode)

<div style="display: flex; justify-content: center">
    <img src="../assets/janela-vscode.png" alt="Visual Studio Code" />
</div>

Embora o **Visual Studio Code**, também conhecido como **VSCode** (não confundir com **Microsoft Visual Studio**) seja, na verdade, um editor de códigos, ele é tão completo que costuma ser tratado como uma verdadeira IDE. Na verdade, é considerado a **IDE mais usada no mundo**. Por esse motivo (e por tantos outros também) esse é o programa de preferência para o curso (mas a escolha do programa pode ficar à critério do professor ou do aluno).

Embora ele já seja nativamente um programa completo, a grande graça do software são as **extensões**, pois elas aumentam a produtividade e o número de recursos disponíveis. Com as extensões certas, o programa pode virar uma IDE tão ou até mesmo mais poderosa que as concorrentes, mesmo sendo 100% gratuíto. Ele é tão completo que é possível utilizar o VSCode para programar em Python em qualquer caminho que seguir: desde programação de aplicativos a até mesmo ciência de dados, ele é simplesmente perfeito.

Ele também é um programa relativamente leve: roda em qualquer máquina, o que o torna ideal para máquinas mais humildes ou mais antigas, caso não tenha dinheiro para comprar um PC mais parrudo. Outra vantagem é que o VSCode também trabalha muito bem com diversas linguagens de programação diferentes, o que o torna ideal para quem for desenvolver um sistema que se utilize mais de uma linguagem.

## PyCharm

<div style="display: flex; justify-content: center">
    <img src="../assets/pycharm.png" alt="PyCharm" />
</div>

O **PyCharm** é a IDE para Python desenvolvimento pela **JetBarins**, empresa essa que está se destacando na criação de IDEs para várias linguagens diferentes. O IntelliJ para o Java é outro exemplo deles.

Mas enfim, a questão é que em se tratando de uma IDE específica para a linguagem Python, o PyCharm se destaca. Extremamente otimizado, o programa possui recursos nativos que tornam o desenvolvimento do software muito mais simplificado que no VSCode, por exemplo.

Podemos citar como destaque a criação de um ambiente virtual e de um repositório Git automático durante a criação de um novo projeto. Outro recursos que é uma mão na roda também é a possibilidade de pesquisar por pacotes externos existentes na Internet e de instalação via interface gráfica.

Possui uma versão **Community**, que é grátis, porém limitada, e uma versão mais completa, e paga: a **Professional**.

## IDLE

<div style="display: flex; justify-content: center">
    <img src="../assets/idle.png" alt="IDLE" />
</div>

O **IDLE** é uma IDE que é instalada automaticamente junto com o Python. Apesar de sua simplicidade, caso não deseje a instalação de softwares adicionais ocupando espaço de armazenamento em disco no seu computador, pode usar este que vem de graça com o Python.

## Jupyter Lab

<div style="display: flex; justify-content: center">
    <img src="../assets/jupyter-lab.png" alt="Jupyter Lab" />
</div>

O **Jupyter Lab** é uma versão mais parruda do já conhecido **Jupyter Notebook**. Embora o Jupyter Notebook possa também ser tratado como uma IDE, ele não possui tantos recursos, já que ele é voltado apenas para Ciência de Dados. E é aí onde entra o Jupyter Lab, que além dos mesmos recursos do Jupyter Notebook, ele também possui ferramentas para desenvolvimento de programas Python.

É totalmente gratuíto, e sua execução ocorre diretamente no navegador de Internet, mesmo que o programa rode localmente.

## Spyder

<div style="display: flex; justify-content: center">
    <img src="../assets/spyder.png" alt="Spyder" />
</div>

O **Spyder** é uma IDE para Python como qualquer outra, mas com um diferencial: ele possui uma visualização em tempo real da saída do algoritmo ao lado da tela do código, mostrando os dados de forma imediata. Ele também possui suporte para muitos recursos voltados para Ciência de Dados, como gráficos e tabelas, por exemplo.

## Orange

<div style="display: flex; justify-content: center">
    <img src="../assets/orange.png" alt="Orange" />
</div>

Orange Data Mining é uma ferramenta open source que permite criar todo fluxo de trabalho de um projeto de data mining, sem necessidade de código. Ideal para quem quer praticar machine learning mas não pretende aprender a codificar ou para iniciantes que pretendem aplicar alguns conceitos, bem como para experts no assunto. Apesar disso, é possível usar scripts Python no próprio Orange.

## Mu

<div style="display: flex; justify-content: center">
    <img src="../assets/mu.png" alt="Mu" />
</div>

A IDE Mu  é voltada para o ensino de programação. Sua aparência simplificada e o fato de possuir uma instalação de Python embutida facilitam bastante o uso por quem está aprendendo a programar.

## Wing

<div style="display: flex; justify-content: center">
    <img src="../assets/wing.png" alt="Wing" />
</div>

Este é um IDE desenvolvido pela Wingware e é especialmente projetado para a linguagem de programação Python. O Wing nos oferece muitos recursos como preenchimento automático, edição automática, navegador de código-fonte, navegação de código e depuração local e remota para que possamos desenvolver nossos programas.

Trata-se de um ambiente de desenvolvimento integrado (IDE) que foi projetado para reduzir o tempo de desenvolvimento e depuração. Ele fornece uma boa ajuda na codificação ou localização de erros. Facilita a navegação e compreensão do código Python.

Assim como o PyCharm, possui uma versão grátis limitada e uma paga mais completa.

## Thonny

<div style="display: flex; justify-content: center">
    <img src="../assets/thonny.png" alt="Thonny" />
</div>

Thonny é um ambiente de desenvolvimento integrado (IDE) que se destaca por sua simplicidade e foco em programadores que estão começando a aventurar-se pela linguagem Python. Com funcionalidades desenhadas para tornar o aprendizado de programação uma experiência descomplicada e estimulante, Thonny oferece uma plataforma sólida para quem deseja mergulhar na codificação sem se perder em configurações e complexidades excessivas.

A premissa básica por trás do Thonny é proporcionar uma interface amigável para novatos na programação. Com uma instalação direta e sem necessidade de ajustar extensões ou parâmetros avançados, o usuário pode começar a codificar quase que imediatamente após o download. A IDE apresenta funcionalidades como execução passo a passo de código e visualização de variáveis em tempo real, o que permite aos usuários entender o que cada linha de código faz e como os dados são manipulados durante a execução do programa.

## Visual Studio IDE for Python

<div style="display: flex; justify-content: center">
    <img src="../assets/visual-studio-ide.png" alt="Visual Studio IDE" />
</div>

Apesar de o **Microsoft Visual Studio** (não confundir com **Visual Studio Code**) ser mais conhecido como uma IDE para a linguagem C#, ele possui uma versão para outras linguagens, incluíndo uma versão para o Python. O Visual Studio é um IDE do Python poderoso no Windows. O Visual Studio é compatível com software livre para a linguagem Python por meio de cargas de trabalho de Desenvolvimento em Python e de Ciência de Dados (Visual Studio 2017 e posteriores) e a extensão gratuita das Ferramentas Python para Visual Studio (Visual Studio 2015 e anteriores).

## Outras alternativas

É claro que essas não são as únicas alternativas para Python. O melhor mesmo é você correr atrás de uma ferramenta que se identifica e usá-la para o desenvolvimento. Não há problemas em usar uma ferramenta alternativa para desenvolver, mas lembre-se: o professor **só dará suporte para a ferramente que ele estiver utilizando**. Caso deseje usar uma ferramenta alternativa, saiba que você deverá conhecer a ferramenta por conta própria.
